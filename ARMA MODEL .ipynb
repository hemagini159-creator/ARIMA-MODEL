```python
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

data = np.array([10, 12, 13, 15, 18, 20, 22, 25, 27, 30])

model = ARIMA(data, order=(1, 0, 1))
results = model.fit()

forecast = results.forecast(steps=5)

print(results.summary())
print("Forecast:", forecast)

# --------- Line Graph ----------
original_x = np.arange(len(data))
forecast_x = np.arange(len(data), len(data) + 5)

plt.plot(original_x, data)
plt.plot(forecast_x, forecast)

plt.xlabel("Time")
plt.ylabel("Values")
plt.title("ARIMA Forecast Line Graph")
plt.legend(["Actual Data", "Forecast Data"])

plt.show()
```
